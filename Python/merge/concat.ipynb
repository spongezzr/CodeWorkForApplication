# Concatenate

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
df1
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                     index=[4, 5, 6, 7])
df2
```

```python
pd.concat([df1, df2])
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']})
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2]).reset_index()
```

```python
pd.concat([df1,df2]).reset_index(drop=True)
```

## attaching keys to the frames

```python
result = pd.concat([df1,df2], keys=['df1', 'df2'])
result
```

```python
# now we can access the data from 'df1':

result.loc['df1']
```

```python
result.iloc[0,1]
```

## Concat in axis=1 direction

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']}, index=[2,3,6,7])
df4
```

```python
df1
```

```python
pd.concat([df1, df4], axis=1)
```

## Ignoring the index

```python
pd.concat([df1, df4], ignore_index=True)
```

## Other variations - example

```python
pd.concat( [df1.reset_index(drop=True), df4.reset_index(drop=True)], axis=1)
```
