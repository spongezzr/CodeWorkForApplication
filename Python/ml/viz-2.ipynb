````python
import seaborn as sns
import numpy as np
%matplotlib inline

sns.set(color_codes=True)
````

````python
x = np.random.normal(size=100)
sns.distplot(x);
````

````python
import pandas as pd

x1 = np.random.randn(100)
x2 = np.random.randn(100)
x3 = np.random.randn(100)

df = pd.DataFrame({'x1':x1, 'x2':x2, 'x3':x3})
corr = df.corr()
````

````python
sns.heatmap(corr)
````
